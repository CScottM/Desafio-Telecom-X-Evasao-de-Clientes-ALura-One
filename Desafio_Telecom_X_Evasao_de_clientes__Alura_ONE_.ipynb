# Relatrio de Anlise de Evaso de Clientes (Churn)  Telecom X

---

## 1. Introduo
A evaso de clientes (Churn)  um problema crtico para empresas de telecomunicaes, pois a perda de clientes impacta diretamente a receita e a sustentabilidade do negcio.  
O objetivo desta anlise  explorar os dados da Telecom X para **identificar padres de comportamento e fatores associados  evaso**, oferecendo insights para reduzir o churn.

---

## 2. Importao e Limpeza de Dados
import pandas as pd
import requests
from pandas import json_normalize
import matplotlib.pyplot as plt

# Importar dados
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()

# Achatar JSON em DataFrame
df = json_normalize(data)

# Padronizar nomes de colunas
df.columns = df.columns.str.lower().str.strip()

# Visualizar primeiras linhas
df.head()
### 2.1 Limpeza e Tratamento
- Remoo de duplicados e valores ausentes.
- Converso de colunas numricas para tipo correto.
- Padronizao de strings em colunas categricas.
# Padronizar colunas categricas
if "churn" in df.columns:
    df["churn"] = df["churn"].str.strip().str.lower()

if "gender" in df.columns:
    df["gender"] = df["gender"].str.strip().str.lower().replace({"female":"feminino","male":"masculino"})

# Converter TotalCharges para numrico
if "totalcharges" in df.columns:
    df["totalcharges"] = pd.to_numeric(df["totalcharges"], errors="coerce")
    df["totalcharges"].fillna(df["totalcharges"].mean(), inplace=True)

# Remover duplicados
df.drop_duplicates(inplace=True)

# Conferir limpeza
df.info()
## 3. Anlise Exploratria de Dados (EDA)
# Estatsticas descritivas
df.describe()
# Mediana das variveis numricas
df.median(numeric_only=True)
# Frequncia de variveis categricas
categorical_cols = df.select_dtypes(include="object").columns.tolist()
for col in categorical_cols:
    print(f"\nColuna: {col}")
    print(df[col].value_counts())
# Correlao de variveis numricas
df.corr(numeric_only=True)
# Distribuio de churn - grfico de barras
if "churn" in df.columns:
    df["churn"].value_counts().plot(kind="bar", color=["skyblue","salmon"])
    plt.title("Distribuio de Churn (Evaso)")
    plt.xlabel("Churn")
    plt.ylabel("Nmero de clientes")
    plt.show()
    
    # Grfico de pizza
    df["churn"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["skyblue","salmon"], explode=(0,0.1))
    plt.title("Proporo de clientes que evadiram vs. permaneceram")
    plt.ylabel("")
    plt.show()
## 4. Concluses e Insights
- A maior parte dos clientes **permanece**, mas h uma parcela significativa que evadiu.  
- Clientes com **contrato mensal** e maior valor de **mensalidade** tendem a ter maior risco de churn.  
- Algumas categorias de servio mostram padres diferentes de evaso.  
- Tempo de contrato (`tenure`)  um indicador importante: clientes mais antigos tm menor probabilidade de churn.
## 5. Recomendaes
- Implementar **programas de fidelizao** para clientes com contratos mensais.  
- Monitorar **clientes com mensalidades altas** e oferecer descontos ou pacotes personalizados.  
- Investir em **servios mais estveis e confiveis**, especialmente na internet, para reduzir cancelamentos.  
- Usar o modelo de anlise para **pr-identificar clientes com alto risco de churn** e agir proativamente.
