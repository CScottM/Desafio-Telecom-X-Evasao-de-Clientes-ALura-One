# Relatório de Análise de Evasão de Clientes (Churn) – Telecom X

---

## 1. Introdução
A evasão de clientes (Churn) é um problema crítico para empresas de telecomunicações, pois a perda de clientes impacta diretamente a receita e a sustentabilidade do negócio.  
O objetivo desta análise é explorar os dados da Telecom X para **identificar padrões de comportamento e fatores associados à evasão**, oferecendo insights para reduzir o churn.

---

## 2. Importação e Limpeza de Dados
import pandas as pd
import requests
from pandas import json_normalize
import matplotlib.pyplot as plt

# Importar dados
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"
response = requests.get(url)
data = response.json()

# Achatar JSON em DataFrame
df = json_normalize(data)

# Padronizar nomes de colunas
df.columns = df.columns.str.lower().str.strip()

# Visualizar primeiras linhas
df.head()
### 2.1 Limpeza e Tratamento
- Remoção de duplicados e valores ausentes.
- Conversão de colunas numéricas para tipo correto.
- Padronização de strings em colunas categóricas.
# Padronizar colunas categóricas
if "churn" in df.columns:
    df["churn"] = df["churn"].str.strip().str.lower()

if "gender" in df.columns:
    df["gender"] = df["gender"].str.strip().str.lower().replace({"female":"feminino","male":"masculino"})

# Converter TotalCharges para numérico
if "totalcharges" in df.columns:
    df["totalcharges"] = pd.to_numeric(df["totalcharges"], errors="coerce")
    df["totalcharges"].fillna(df["totalcharges"].mean(), inplace=True)

# Remover duplicados
df.drop_duplicates(inplace=True)

# Conferir limpeza
df.info()
## 3. Análise Exploratória de Dados (EDA)
# Estatísticas descritivas
df.describe()
# Mediana das variáveis numéricas
df.median(numeric_only=True)
# Frequência de variáveis categóricas
categorical_cols = df.select_dtypes(include="object").columns.tolist()
for col in categorical_cols:
    print(f"\nColuna: {col}")
    print(df[col].value_counts())
# Correlação de variáveis numéricas
df.corr(numeric_only=True)
# Distribuição de churn - gráfico de barras
if "churn" in df.columns:
    df["churn"].value_counts().plot(kind="bar", color=["skyblue","salmon"])
    plt.title("Distribuição de Churn (Evasão)")
    plt.xlabel("Churn")
    plt.ylabel("Número de clientes")
    plt.show()
    
    # Gráfico de pizza
    df["churn"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["skyblue","salmon"], explode=(0,0.1))
    plt.title("Proporção de clientes que evadiram vs. permaneceram")
    plt.ylabel("")
    plt.show()
## 4. Conclusões e Insights
- A maior parte dos clientes **permanece**, mas há uma parcela significativa que evadiu.  
- Clientes com **contrato mensal** e maior valor de **mensalidade** tendem a ter maior risco de churn.  
- Algumas categorias de serviço mostram padrões diferentes de evasão.  
- Tempo de contrato (`tenure`) é um indicador importante: clientes mais antigos têm menor probabilidade de churn.
## 5. Recomendações
- Implementar **programas de fidelização** para clientes com contratos mensais.  
- Monitorar **clientes com mensalidades altas** e oferecer descontos ou pacotes personalizados.  
- Investir em **serviços mais estáveis e confiáveis**, especialmente na internet, para reduzir cancelamentos.  
- Usar o modelo de análise para **pré-identificar clientes com alto risco de churn** e agir proativamente.
